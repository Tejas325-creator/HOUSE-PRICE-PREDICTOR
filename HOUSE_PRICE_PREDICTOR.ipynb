RANDOM FOREST REGRESSION FOR BEST ACCURACY SCORE


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import OneHotEncoder  # Import OneHotEncoder

# Assuming 'housing.csv.zip' is in the current directory
data = pd.read_csv("housing.csv.zip")  # Load the data here to ensure it's available

# One-hot encode 'ocean_proximity'
# Create a OneHotEncoder object
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
# Fit the encoder on the 'ocean_proximity' column
encoded_data = encoder.fit_transform(data[['ocean_proximity']])
# Create a DataFrame from the encoded data
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['ocean_proximity']))
# Concatenate the encoded features with the original DataFrame
data = pd.concat([data, encoded_df], axis=1)
# Drop the original 'ocean_proximity' column
data = data.drop('ocean_proximity', axis=1)

# Now, proceed with feature selection and model training
x = data.iloc[:, :-1].values  # Select all columns except 'median_house_value'
y = data.iloc[:, -1].values  # Select 'median_house_value' as the target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(x_train, y_train)

y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))

print("R-squared:", r2_score(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
forest  = RandomForestRegressor()
param_grid = {
    "n_estimators" : [3,10,30],
    "max_features" : [2,4,6,8]
}
grid_search = GridSearchCV(estimator = forest, param_grid = param_grid, cv = 10)
```
